```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Growth of $100
growth = pd.read_csv("../data/processed/growth_of_100.csv")
growth["date"] = pd.to_datetime(growth["date"])

plt.figure(figsize=(12, 6))
for col in growth.columns[1:]:
    plt.plot(growth["date"], growth[col], label=col)

plt.title("Growth of $100 Across Asset Classes")
plt.xlabel("Date")
plt.ylabel("Portfolio Value")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
corr = pd.read_csv("../data/processed/correlation_matrix.csv", index_col=0)

plt.figure(figsize=(8, 6))
im = plt.imshow(corr, aspect="auto")
plt.colorbar(im)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.index)), corr.index)
plt.title("Correlation Matrix of Monthly Returns")
plt.tight_layout()
plt.show()
```

```python
# Rolling volatility
rolling_vol = pd.read_csv("../data/processed/rolling_vol_12m.csv")
rolling_vol["date"] = pd.to_datetime(rolling_vol["date"])

plt.figure(figsize=(12, 6))
for col in ["AEQ", "ILE_H", "ILE_UH", "AFI", "CASH"]:
    plt.plot(rolling_vol["date"], rolling_vol[col], label=col)

plt.title("12-Month Rolling Annualised Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Asset availability chart
missingness = pd.read_csv("../data/processed/missingness_report.csv")
missingness["first_valid_date"] = pd.to_datetime(missingness["first_valid_date"])
missingness["last_valid_date"] = pd.to_datetime(missingness["last_valid_date"])

plt.figure(figsize=(10, 5))
for i, row in missingness.iterrows():
    plt.plot(
        [row["first_valid_date"], row["last_valid_date"]],
        [row["asset_code"], row["asset_code"]],
        linewidth=6
    )

plt.title("Asset Data Availability")
plt.xlabel("Date")
plt.ylabel("Asset")
plt.tight_layout()
plt.show()
```
